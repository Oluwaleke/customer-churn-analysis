```python
import pandas as pd

df = pd.read_csv("telco_churn.csv")

print(df.shape)
print(df.head())
print(df.info())
```

## Data Cleaning

```python
df.isnull().sum()
```

```python
#Although it looks numeric, TotalCharges is often read as a string due to empty or whitespace values.
print(df['TotalCharges'].dtype)
print(df['TotalCharges'].head(10))
```

```python
#convert it safely
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
#checking for any NaNs introduced
df[df['TotalCharges'].isna()]
```

```python
#They’re few (e.g., just customers with tenure == 0),so we can drop them
df.dropna(inplace=True)
```

```python
#Checking for inconsistent categorical values... This helps know how to group, visualize, or encode them later.
print(df['Churn'].unique())
print(df['Contract'].unique())
print(df['PaymentMethod'].unique())
```

```python
#Dropping customer ID since It’s just an identifier and not useful for analysis.
df.drop(columns='customerID', inplace=True)
```

## Exploratory Data Analysis (EDA)

```python
#Set up visual tools 
import seaborn as sns
import matplotlib.pyplot as plt

# Style setup
sns.set(style="whitegrid")
```

#### 1. Churn by Gender

```python
sns.countplot(data=df, x='gender', hue='Churn')
plt.title('Churn Rate by Gender')
plt.show()
```

#### 2. Churn by Senior Citizen

```python
sns.countplot(data=df, x='SeniorCitizen', hue='Churn')
plt.title('Churn Rate by Senior Citizen Status')
plt.xticks([0, 1], ['Not Senior', 'Senior'])
plt.show()
```

#### 3. Churn by Contract Type

```python
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Churn Rate by Contract Type')
plt.xticks(rotation=15)
plt.show()
```

#### 4. Churn vs Monthly Charges

```python
sns.boxplot(data=df, x='Churn', y='MonthlyCharges')
plt.title('Monthly Charges vs Churn')
plt.show()
```

#### 5. Churn by Internet Service

```python
sns.countplot(data=df, x='InternetService', hue='Churn')
plt.title('Churn Rate by Internet Service Type')
plt.xticks(rotation=15)
plt.show()
```

#### 6. Tenure Distribution

```python
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', bins=30)
plt.title('Tenure Distribution by Churn Status')
plt.xlabel('Customer Tenure (Months)')
plt.ylabel('Number of Customers')
plt.show()
```

## Business Insights & Recommendations.

### ✅ **Insights**

1. **Contract Type Drives Churn**
   Customers on **month-to-month contracts** churn significantly more than those on **1-year or 2-year contracts**. Long-term contracts appear to be more effective in retaining customers.

2. **Higher Monthly Charges Increase Churn Risk**
   Customers with **higher monthly charges** are more likely to leave. This suggests price sensitivity among high-paying customers

3. **Low Tenure, High Churn**
   Most churned customers are in their **first few months of service**. Early retention efforts could drastically reduce churn

4. **Senior Citizens Are More Likely to Churn**
   The churn rate is higher among **senior citizens**, possibly due to usability or support concerns.5. **No Internet = Less Churn**
   Surprisingly, customers with **no internet service** churn the least. This could mean less complexity in their service needs.
.




### ✅ **Recommendations**

* Promote **annual contract upgrades** with incentives to shift customers from monthly plans.
* Introduce **welcome offers or personalized onboarding** to improve the first-3-month experience.
* Monitor and support **high-spending customers** with loyalty perks or check-ins.
* Tailor **support for seniors**, possibly through better customer service or simplified plans.
* Explore upsell strategies for non-internet customers who are stable and loyal.


### ✅ **Conclusion**

This analysis of Telco customer churn revealed key service and demographic patterns that contribute to customer loss. By targeting new customers, senior citizens, and those on monthly contracts, the company can take strategic steps to reduce churn and improve customer lifetime value.
